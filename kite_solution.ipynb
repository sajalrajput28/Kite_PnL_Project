```python
def calculate_charges(df):
    
    df = df.copy()
    
    # Turnover
    df['Turnover'] = df['Qty'] * df['Avg. price']

    #Brokreage
    df['Brokerage'] = np.minimum(df['Turnover'] * 0.0003,20)

    #STT/CTT
    df['STT/CTT'] = np.where(df['Type']=='SELL', df['Turnover'] * 0.00025,0)

    #ETC
    df['ETC'] = df['Turnover'] * 0.0000297

    #SEBI
    df['SEBI'] = df['Turnover'] * 10/10000000

    #GST
    df['GST'] = (df['Brokerage']+df['SEBI']+df['ETC']) * 0.18

    #STAMP_CHARGES
    df['Stamp_charges'] = np.where(df['Type']=='BUY', df['Turnover'] * 0.00003,0)

    #TOTAL_CHARGES
    df['Total_charges'] = df['Brokerage']+df['STT/CTT']+df['ETC']+df['SEBI']+df['GST']+df['Stamp_charges']
    
    return df
    
```

```python
df = calculate_charges(df)
df.head()
```

```python
stock_type_wise = df.groupby(['Instrument','Type']).agg(
    Total_qty = ('Qty','sum'),
    Weighted_avg_price = ('Avg. price', lambda x:np.average(x,weights=x)),
    Total_turnover = ('Turnover','sum'),
    Total_brokerage = ('Brokerage','sum'),
    Total_stt_ctt = ('STT/CTT','sum'),
    Total_etc = ('ETC','sum'),
    Total_sebi = ('SEBI','sum'),
    Total_gst = ('GST','sum'),
    Total_stamp = ('Stamp_charges','sum'),
    Total_Charges = ('Total_charges','sum')
).reset_index()
stock_type_wise
```

3) Overall Summary of each Stocks 

TABLE 3

```python
stock_type_wise
```

```python
stock_type_wise['Buy Value']=np.where(stock_type_wise['Type']=='BUY',stock_type_wise['Total_turnover'],0)
stock_type_wise['Sell Value']=np.where(stock_type_wise['Type']=='SELL',stock_type_wise['Total_turnover'],0)
stock_type_wise['PnL']=stock_type_wise['Sell Value']-stock_type_wise['Buy Value']
df = stock_type_wise
```

```python
df
```

```python
stock_summary=df.groupby('Instrument').agg(Gross_PnL=('PnL','sum'),Total_Charges=('Total_Charges','sum')).reset_index()
stock_summary
```

```python
# calculate net profit and loss(PnL) 
stock_summary['Net_PnL']=stock_summary['Gross_PnL']-stock_summary['Total_Charges']
# calculate the percentage of charges on Gross PnL
stock_summary['% Charges on Gross PnL']=(stock_summary['Total_Charges'] / stock_summary['Gross_PnL'])*100
stock_summary
```
